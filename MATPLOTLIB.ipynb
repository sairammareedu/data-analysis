```python
tigar=np.linspace(-np.pi,np.pi,100)
plt.plot(tigar,np.sin(tigar),color='black',linestyle='dotted',label='sin(x)')
plt.title('sin(x)')
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
runs = np.array([100,50,91,78,89,25,34,19,9,10])
w = np.array([1,0,2,0,3,7,8,9,7,5])
plt.plot(w,runs,color='red',label='Ind')
#plt.plot(x,y,color,label.....)
plt.legend(loc='best')
plt.title('IndvsAus_score')
plt.show()
```

```python
a = [230,560,780,127,128]
b = [200,160,270,127,400]
years = [1,2,3,4]
profit_a = [(a[i]-a[i-1]) for i in range(1,len(a))]
profit_b = [(b[i]-b[i-1]) for i in range(1,len(b))]
plt.subplot(2,1,2)
plt.plot(years,profit_a,color='purple',linewidth = '1',label =
'CompanyA',marker='H',ms='10',mec='c')
plt.subplot(2,1,1)
plt.plot(years,profit_b,color='black',linestyle='dotted',label =
'CompanyB',marker = 'H')
plt.show()
```

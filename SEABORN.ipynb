```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
sns.scatterplot(x="total_bill",y="tip",data=tips,label='tips')
plt.title("Scatter plot of Total Bill vs Tip")
plt.xlabel("Total Bill($)")
plt.ylabel("Tip($)")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Load example dataset
tips=sns.load_dataset("tips")
print(tips)
#create a scatter plot
sns.jointplot(x="day",y="total_bill",data=tips,label='tips')
plt.legend(loc='best')
plt.title("scatter plot of Day of the week vs. Total bill")
plt.xlabel("Day of the week($)")
plt.ylabel("Total bill($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#Load example dataset
tips=sns.load_dataset("tips")
print(tips)
#create a scatter plot
sns.countplot(x="day",data=tips,label='tips')
plt.legend(loc='best')
plt.title("scatter plot of Day of the week vs. Total bill")
plt.xlabel("Day of the week($)")
plt.ylabel("Total bill($)")
plt.show()
```

```python
tips=sns.load_dataset("tips")
sns.lmplot(x="tip",y="total_bill",data=tips)
plt.title("scatter plot of Day of the week vs. Total bill")
plt.xlabel("Tips($)")
plt.ylabel("Total bill($)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
g=sns.FacetGrid(titanic,col="class")
g.map(sns.histplot,"age")
plt.subplots_adjust(top=0.8)
g.fig.suptitle("Age Distribution By passenger class")
plt.show()
```
